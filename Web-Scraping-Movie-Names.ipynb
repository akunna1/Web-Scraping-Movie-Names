```python
import requests
from bs4 import BeautifulSoup

# Defining the base URL
base_url = "https://subslikescript.com/movies?page="

# Iterating through pages 1 to 10
page_number = 1
while page_number <= 10:
    # Constructing the URL for the current page
    url = base_url + str(page_number)
    
    # Making a GET request to the specified URL
    response = requests.get(url)
    
    # Parsing the content of the response using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Locating the "ul" element with class "scripts-list" and storing it in a variable
    script_list = soup.find('ul', class_="scripts-list")
    
    # Check if script_list is found
    if script_list:
        # Looping through all the "a" elements inside the "ul" element and printing the text attribute
        for a in script_list.find_all('a'):
            print(a.text)
    else:
        print(f"No script list found on page {page_number}")
    
    # Move to the next page
    page_number += 1
```
